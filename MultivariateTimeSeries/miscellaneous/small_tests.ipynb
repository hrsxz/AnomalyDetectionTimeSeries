### Plot sine wave

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义数据点
x = np.linspace(0, 200 * np.pi, 10000)  # 生成从0到2π的100个点
y = np.sin(x)  # 计算每个点的正弦值

# 绘制正弦曲线
plt.figure(figsize=(8, 4))  # 设置图形大小
plt.plot(x, y, label='sin(x)')  # 绘制曲线
plt.title('Sinus Curve')  # 添加标题
plt.xlabel('x')  # x轴标签
plt.ylabel('sin(x)')  # y轴标签
plt.legend()  # 显示图例
plt.grid(True)  # 显示网格
plt.show()  # 显示图形
```

### While循环测试

```python
import time

machine_enabled = False
wait_for_machine_enable = True
last_time = time.time()
        
while False:
    time.sleep(1)
    print("Global while loop running...") 
    print("Gathering data...")

    while not machine_enabled and wait_for_machine_enable:
        time.sleep(1)
        print("Read machine ready status...")
        machine_enabled = False
        if machine_enabled:
            print("Machine is enabled.")
            wait_for_machine_enable = False
            break
    else:
        print("read machine ready status...")
        if not machine_enabled:
            print("Machine is not enabled.")
            wait_for_machine_enable = True
    
    if not machine_enabled:
        break

    print("Pushing data in queue...")
```
