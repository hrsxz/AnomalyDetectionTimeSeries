### Plot sine wave

```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.linspace(0, 200 * np.pi, 10000)  # 02100
y = np.sin(x)  # 

# 
plt.figure(figsize=(8, 4))  # 
plt.plot(x, y, label='sin(x)')  # 
plt.title('Sinus Curve')  # 
plt.xlabel('x')  # x
plt.ylabel('sin(x)')  # y
plt.legend()  # 
plt.grid(True)  # 
plt.show()  # 
```

### While

```python
import time

machine_enabled = False
wait_for_machine_enable = True
last_time = time.time()
        
while False:
    time.sleep(1)
    print("Global while loop running...") 
    print("Gathering data...")

    while not machine_enabled and wait_for_machine_enable:
        time.sleep(1)
        print("Read machine ready status...")
        machine_enabled = False
        if machine_enabled:
            print("Machine is enabled.")
            wait_for_machine_enable = False
            break
    else:
        print("read machine ready status...")
        if not machine_enabled:
            print("Machine is not enabled.")
            wait_for_machine_enable = True
    
    if not machine_enabled:
        break

    print("Pushing data in queue...")
```
